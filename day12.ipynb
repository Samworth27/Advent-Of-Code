```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
file = open('day12.txt')
heightmap = []
start = (0, 0)
finish = (0, 0)
y = 0
for line in file:
    x = 0
    new_line = []
    for char in line.strip():
        if char == 'S':
            start = (x, y)
            char = 'a'
        if char == 'E':
            finish = (x, y)
            char = 'z'
        new_line.append(ord(char)-97)
        x += 1
    heightmap.append(new_line)
    y += 1

sns.heatmap(heightmap)
plt.axis('equal')
```
